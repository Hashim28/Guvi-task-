# Groupby and rolling methods can be used to find the users who were active for more than 3 days in any of the 7 day rolling period.

```python
df = users_engagement.copy()
```

```python
df
```

```python
df['visited_7_days'] = df.groupby('user_id',as_index = False,group_keys = False).apply(lambda x:x.rolling('7D',on = 'time_stamp')['user_id'].count())
```

```python
df
```

If any of the values in visited_7_days column is greater than 2, then it is saved in adopted user column

```python
user_adopted = df.groupby('user_id')['visited_7_days'].max().reset_index()
```

```python
user_adopted.head()
```

```python
user_adopted['adopted_user'] = user_adopted['visited_7_days'].apply(lambda x: 1 if x >= 3 else 0)
```

```python
user_adopted.head()
```

```python
user_adopted['adopted_user'].value_counts()
```

# Joining user_adopted to users dataframe

```python
user_adopted.drop('visited_7_days',axis = 1, inplace = True)
```

```python
user_adopted.rename(columns = {'user_id':'object_id'},inplace = True)
```

```python
user_adopted.set_index('object_id',inplace = True)
```

```python
user_adopted
```

```python
df_users = users.join(user_adopted, on = 'object_id', how = 'left')
```

```python
df_users
```

```python
df_users.info()
```

The null values in the _adopted_user and _last_session_creation_time can be filled in with 0 by assuming that those users are not adopted users.

```python
df_users['last_session_creation_time'].fillna(0, inplace = True)
df_users['adopted_user'].fillna(0, inplace = True)
```

Dropping columns which are not useful.
